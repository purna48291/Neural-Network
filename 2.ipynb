defining a empty string

```python
st = ""
```

defining a function to count occurances of each word

```python
def wordcount(st):
    counts = dict()
    w = st.split()
    for y in w:
        if y in counts:
            counts[y]+=1
        else:
            counts[y]=1
    return counts
```

reading input.txt file and adding it to st variable

```python
with open('/content/drive/MyDrive/ICP_1/input.txt','r') as file:
    data = file.read()
    st +=data
```

calling wordcount function

```python
a=str(wordcount(st))
st+="\nWord_Count: \n"
st+=a
```

replacing string to get required output

```python
st = st.replace("{","")
st = st.replace("}","")
st = st.replace("'","")
st = st.replace("P","\nP")
st = st.replace(", ","\n")
print(st)
```

opening and writing to the file

```python
outputfile = open('/content/drive/MyDrive/ICP_1/output.txt.ipynb','w')
outputfile.write(st)
```

closing of txt files

```python
outputfile.close()
file.close()
```
